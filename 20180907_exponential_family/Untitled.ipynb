# Exponential family distribution 的に考える Multivariate Normalモデルと共役事前分布

Lin (2013)を読んでいると, exponential family distribution と共役事前分布について下のような式が登場した.

$$
\begin{eqnarray}
\mu \left(d \theta \mid \lambda, \lambda^\prime \right) \tag{17}
&=&
    \exp \left(
        \lambda^{\rm T} \eta \left( \theta \right)
        - \lambda^\prime A\left( \theta \right)
        -B\left( \lambda, \lambda^\prime\right)
    \right) h\left( d\theta \right) \\
F\left( x \mid \theta \right) \tag{18}
&=&
    \exp \left(
        \eta\left( \theta \right) ^{\rm T} T \left( x \right)
        - \tau A \left(\theta \right)
        - b\left( x \right)
    \right) \\
\int_\theta F \left( x \mid \theta \right) \mu \left( d\theta \mid \lambda, \lambda^\prime\right) \tag{19}
&=&
    \exp \left(
        B\left( \lambda + T\left(x \right), \lambda^\prime + \tau \right)
        - B\left( \lambda, \lambda^\prime\right)
        -b\left( x \right)
    \right)
\end{eqnarray}
$$


ここで, $F$ は確率測度, $\mu$ は $F$ の共役事前分布. $(\lambda, \lambda^\prime) $ は $\mu$ のnatural parameters である. $x_1, ..., x_n $ のデータが得られたときの　$\mu $ の事後分布の natural parameters は $(\lambda + \sum_{i=1}^n T(x_i), \lambda^\prime + n\tau) $ となる.

$ F $ が exponential family distribution であるときに上が一般に成り立つ, と書いてある.

意味がわからないので具体的な分布について考えてみる. $ F $ を多変量正規分布, $\mu$ をNormal-Wishart分布とする.

## 多変量正規分布の場合

つまり, 以下の形式に変形する.

- expの中にすべて格納.
- expの中身は, パラメータ $ \theta $ の関数と $ x $ の関数の積, $ \theta $ の関数, $ x $ の関数, の和

多変量正規分布の確率密度関数がこちら.

$$
\begin{eqnarray}
\mathcal N \left(\textbf x \mid \boldsymbol \mu, \boldsymbol \Lambda^{-1} \right)
&=&
    \frac {\left| \boldsymbol \Lambda \right|^\frac {1} {2}} {(2\pi)^{\frac{D}{2}}}
    {\rm exp }\left(
        - \frac {1} {2}
        \left(\textbf x - \boldsymbol \mu\right)^{\rm T}
        \boldsymbol \Lambda
        \left(\textbf x - \boldsymbol \mu\right)
    \right)
\end{eqnarray}
$$

$ \boldsymbol \Lambda $ が対称行列であること, $\mathbf x^{\rm T} \boldsymbol \Lambda \boldsymbol \mu $ はスカラ量なので$\mathbf x^{\rm T} \boldsymbol \Lambda \boldsymbol \mu = \operatorname {Tr} \left( \mathbf x^{\rm T} \boldsymbol \Lambda \boldsymbol \mu \right)$ が成り立つことから以下のようにしてexponential family に変形できる.


$$
\begin{eqnarray}
\frac {\left| \boldsymbol \Lambda \right|^\frac {1} {2}} {(2\pi)^{\frac{D}{2}}}
{\rm exp }\left(
    - \frac {1} {2}
    \left(\textbf x - \boldsymbol \mu\right)^{\rm T}
    \boldsymbol \Lambda
    \left(\textbf x - \boldsymbol \mu\right)
\right)
&=&
    \frac {\left| \boldsymbol \Lambda \right|^\frac {1} {2}} {(2\pi)^{\frac{D}{2}}}
    {\rm exp }\left(
        - \frac {1} {2} \textbf x^{\rm T} \boldsymbol \Lambda \textbf x
        + \frac {1} {2} \textbf x^{\rm T} \boldsymbol \Lambda \boldsymbol \mu
        + \frac {1} {2} \boldsymbol \mu^{\rm T} \boldsymbol \Lambda \textbf x
        - \frac {1} {2} \boldsymbol \mu^{\rm T} \boldsymbol \Lambda \boldsymbol \mu
    \right) \\
&=&
    \frac {\left| \boldsymbol \Lambda \right|^\frac {1} {2}} {(2\pi)^{\frac{D}{2}}}
    {\rm exp }\left(
        - \frac {1} {2} \operatorname {Tr} \left( \textbf x^{\rm T} \boldsymbol \Lambda \textbf x \right)
        + \frac {1} {2} \textbf x^{\rm T} \left( \boldsymbol \mu^{\rm T} \boldsymbol \Lambda^{\rm T} \right)^{\rm T}
        + \frac {1} {2} \boldsymbol \mu^{\rm T} \boldsymbol \Lambda \textbf x
        - \frac {1} {2} \boldsymbol \mu^{\rm T} \boldsymbol \Lambda \boldsymbol \mu
    \right) \\
&=&
    \frac {\left| \boldsymbol \Lambda \right|^\frac {1} {2}} {(2\pi)^{\frac{D}{2}}}
    {\rm exp }\left(
        - \frac {1} {2} \operatorname {Tr} \left(\boldsymbol \Lambda \textbf x \textbf x^{\rm T} \right)
        + \frac {1} {2} \left( \boldsymbol \mu^{\rm T} \boldsymbol \Lambda \textbf x \right)^{\rm T}
        + \frac {1} {2} \boldsymbol \mu^{\rm T} \boldsymbol \Lambda \textbf x
        - \frac {1} {2} \boldsymbol \mu^{\rm T} \boldsymbol \Lambda \boldsymbol \mu
    \right) \\
&=&
    \frac {\left| \boldsymbol \Lambda \right|^\frac {1} {2}} {(2\pi)^{\frac{D}{2}}}
    {\rm exp }\left(
        - \frac {1} {2} \boldsymbol \Lambda \textbf x \textbf x^{\rm T}
        + \boldsymbol \mu^{\rm T} \boldsymbol \Lambda \textbf x
        - \frac {1} {2} \boldsymbol \mu^{\rm T} \boldsymbol \Lambda \boldsymbol \mu
    \right) \\
&=&
    {\rm exp }\left(
        - \frac {1} {2} \boldsymbol \Lambda \textbf x \textbf x^{\rm T}
        + \left( \boldsymbol \Lambda \boldsymbol \mu\right)^{\rm T} \textbf x
        - \frac {1} {2} \boldsymbol \mu^{\rm T} \boldsymbol \Lambda \boldsymbol \mu
        + \ln \left(\frac {\left| \boldsymbol \Lambda \right|^\frac {1} {2}} {(2\pi)^{\frac{D}{2}}} \right)
    \right) \\
&=&
    {\rm exp }\left(
        \left(
            \begin{array} {cc}
            \boldsymbol \Lambda \boldsymbol \mu \\
            -\frac {1} {2} \boldsymbol \Lambda \\
            \end{array}
        \right)^{\rm T}
        \left(
            \begin{array} {cc}
            \mathbf x \\
            \mathbf x\mathbf x^{\rm T}
            \end{array}
        \right)
        - \frac {1} {2} \boldsymbol \mu^{\rm T} \boldsymbol \Lambda \boldsymbol \mu
        + \frac {1} {2} \ln \left| \boldsymbol \Lambda \right|
        - \frac{D}{2} \ln (2\pi)
    \right) \\
\end{eqnarray}
$$

$$
\begin{eqnarray}
\eta\left( \theta \right)
&=&
    \left(\begin{array} {cc}
        \boldsymbol \Lambda \boldsymbol \mu \\
        -\frac {1} {2} \boldsymbol \Lambda \\
    \end{array}\right) \\
T\left( x \right)
&=&
    \left(\begin{array} {c}
        \mathbf x \\
        \mathbf x \mathbf x^{\rm T} \\
    \end{array}\right) \\
\tau A\left( \theta \right)
&=&
    \frac {1} {2} \left(
        \boldsymbol \mu^{\rm T} \boldsymbol{ \Lambda \mu}
        - \ln \left| \boldsymbol \Lambda \right|
    \right)\\
b\left( x \right)
&=&
    \frac{D}{2} \ln (2\pi)
\end{eqnarray}
$$

## Wishart分布の場合

$$
\begin{eqnarray}
\mathcal W \left(\boldsymbol \Lambda \mid \textbf W^{-1}, \nu \right)
&=&
{\frac {\left|\boldsymbol \Lambda  \right|^{\frac {\nu-D-1}{2}}{\rm exp} \left(
    \frac
    {-\operatorname {Tr} (\mathbf {W} ^{-1}{\boldsymbol \Lambda })} {2}\right)}
    {2^{\frac {\nu D}{2}}\left|{\mathbf {W} }\right|^{\frac {\nu} {2}}\Gamma _{D}\left({\frac {\nu}{2}}\right)}}
\end{eqnarray}
